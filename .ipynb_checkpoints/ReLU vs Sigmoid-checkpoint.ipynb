## How do ReLU and Sigmoid relate to eachother?

ReLU (Rectified Linear Units) have recently become an alternative activation function to the sigmoid function in neural networks.

ReLUs overfit more easily than sigmoids, this makes them a natural fit for dropout which is a technique to avoid overfitting. ReLUs also train faster than sigmoids due to less numerical computation. ReLUs also don't suffer from the vanishing gradient problem.

In this notebook we'll graphically explore the relationship between the ReLU function and the Sigmoid function.

```python
import numpy as np
import pandas as pd
%matplotlib inline
from ggplot import *
```

```python
# Define the functions
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

@np.vectorize
def relu(x):
    if x < 0:
        return 0.0
    else:
        return x
```

```python
x = np.linspace(-10, 10, 200) # the x axis values
```

```python
d = {'sigmoid': sigmoid(x), 'relu': relu(x), 'x':x}
df = pd.DataFrame(data=d)
df.head()
```

Let's plot both graphs

```python
ggplot(aes(x='x', y='sigmoid'), data=df) + geom_line() + ggtitle('Sigmoid') + ylab('sigmoid(x)')
```

```python
ggplot(aes(x='x', y='relu'), data=df) + geom_line() + ggtitle('ReLU') + ylab('relu(x)')
```

```python
p = ggplot(aes(x='x', y='relu'), data=df) + geom_line() + geom_line(aes(x='x', y='sigmoid'), data=df)
p = p + ggtitle('Sigmoid vs ReLU') + ylab('sigmoid(x) vs relu(x)')
p
```

The relu function keeps increasing while the sigmoid function plateaus at 1. For the purposes of the visualiations we'll only pay attention to relu(x) >=0 and <= 1.0.

It's not clear at this point how relu relates to sigmoid.

```python
offsets = np.linspace(0, 10, 20)
# Blue line is the one with no offset
p = ggplot(aes(x='x',y='sigmoid'), data=df) + geom_line(color='blue') + ggtitle('Sigmoid Offsets')
for offset in offsets:
    p = p + geom_line(aes(x='x', y=sigmoid(x-offset+0.5)), data=df)
p
```

This still doesn't really show anything relu related but it's important to note that this illustrates many different inputs and outputs of the sigmoid function.

What if we summed it all?

```python
avg = sigmoid(x)
for offset in offsets:
    so = sigmoid(x-offset+0.5)
    avg += so
```

```python
ggplot(aes(x='x', y=avg), data=df) + geom_line() + ggtitle('Summed Sigmoid')
```

This is starting to look like the relu function!

In fact if we take sum(sigmoid(x-offset+0.5) for offset in range(1, infinity)) we get roughly log(exp(x) + 1) since we're in the discrete case.

```python
ggplot(aes(x='x', y=np.log(np.exp(x)+1)), data=df) + geom_line()
```

For simplicity and numerical optimization purposes we then just approximate log(exp(x) + 1) as max(x, 0).

Therefore ReLU(x) is max(x, 0).

It's interesting how two seemingly different functions have a strong relation to eachother. It's also comforting that max(x, 0) wasn't just pulled out of a hat!
